```python
#roi 좌표가 여러 개일 때 txt 파일 만들기
import pandas as pd
import glob
import cv2
import matplotlib.pyplot as plt
import os

roi=[]
num=0

input_file = '../rawdata/cervigram_data_A.xlsx'
read_file = pd.read_excel(input_file, engine='openpyxl')
# roi.append(read_file)
# print(len(read_file))

f = open('../data/image/%s/train_A.txt'%num,'w')

for i in range(len(read_file)):
    name = read_file['fcode'][i]
    img_path = './rawdata/img_all/'+str(name) + '.jpg'
#     print(os.path.exists(img_path))

    string = read_file['roi'][i]
    box_list = string.split('/')
    
    for j in range(len(box_list)):
        roi_list = box_list[j].split(',')
        x1 = int(roi_list[1])
        y1 = int(roi_list[2])
        x2 = int(roi_list[3])
        y2 = int(roi_list[4])
        
        print(img_path, x1, y1, x2, y2)
    
        txt = '%s,%s,%s,%s,%s,A\n'%(img_path, x1, y1, x2, y2)
        f.write(txt)
        
f.close()
```
